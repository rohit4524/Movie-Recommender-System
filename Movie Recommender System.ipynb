```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on = 'title') # Merging both dataset to simplify things
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
# Pre - Processing of data to keep only useful columns that will help us in filtering.
# Genres, id, keywords, , overview, title, cast, crew

movies = movies[['genres', 'id', 'keywords', 'overview', 'title', 'cast', 'crew']]
```

```python
movies.head(1) #Final Data
```

```python
movies.shape
```

```python
# Pre - Processing our final data like removing missing, duplicates data, etc

movies.isnull().sum()                         # Missing Data
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()                                 # Duplicates
```

```python
movies.shape                 
```

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# Pre - Processing of genres column since the given format is not good (Dictionary).

def change(x):
    List = []
    
    for i in ast.literal_eval(x):
        List.append(i['name'])
    return List
```

```python
movies['genres'].head()
```

```python
 movies.head()
```

```python
movies['genres'] = movies['genres'].apply(change)
```

```python
movies['keywords'] = movies['keywords'].apply(change)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
# Pre Processing of the Cast Column
# Getting the names of only Leading Cast with their original names, not their Character name

def change_cast(x):
    List = []
    counter = 0
    for i in ast.literal_eval(x):
        if counter != 3:
            List.append(i['name'])
            counter += 1
        else:
            break
    return List
```

```python
movies['cast'] = movies['cast'].apply(change_cast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
# Pre - Processing of Cast Column  
# Will Fetch all the Director name only

def fetch_director(x):
    List = []
    
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            List.append(i['name'])
            break
    return List
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
# Making List of Sentences in Overview Column

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

movies.head()

```python
movies.head()
```

```python
# Removing space character to make all entities unique

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# Merging columns into one

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
final = movies[['id', 'title', 'tags']]
```

```python
final.head()
```

```python
final['tags'] = final['tags'].apply(lambda x:" ".join(x))
```

```python
final['tags'][0]
```

```python
final['tags'] = final['tags'].apply(lambda x:x.lower())
```

```python
#Lemmatization of Processed Data

import nltk
from nltk.corpus import wordnet
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lem(text):
    y = []
    
    for i in text.split():
        y.append(lemmatizer.lemmatize(i, wordnet.VERB))
        
    return " ".join(y)
```

```python
final['tags'] = final['tags'].apply(lem)
```

```python
# Text Vectorization for Gettting Similarity 

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
cv.fit_transform(final['tags']).toarray()    #Converting to Vectors
```

```python
cv.fit_transform(final['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(final['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()                         # After Lemmatization
```

```python
# Cosine Similarity 

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Passing all the vectors

cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[2]
```

# Functions for Recommending similar movie

```python
# Getting Index Of the Input Movie

final[final['title'] == 'Avatar'].index[0]
```

```python
# Getting Distance of input movie with all the movies

similarity[final[final['title'] == 'Avatar'].index[0]]
```

```python
# Sorting of Distance Vectors

sorted(similarity[0], reverse = True)
```

```python
# To Maintain the Index position while and after Sorting

list(enumerate(similarity[0]))
```

```python
# Sorting the list based on similarity score

sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
# Combining all the Process

def recommend(movie):
    movie_index = final[final['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    
    for i in movies_list:
        print(final.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
